Word based tokenisation


```python
s = 'I like this course'
```

```python
s.split()  # it uses white spaces to split
```

Character based tokenisation

```python
list = []

for char in s:
  list.append(char)

print(list)
```

removing white spaces

```python
text = "".join(s.split())
print(text)
```

```python
list2 = []
for char in text:
  list2.append(char)

print(list2)
```

Subword Tokenisation

```python
string = "I am learning Tokenisation"
```

```python
newstring = string.split()
print(newstring)
```

```python
list3 = []
for word in newstring:
  if word[-3:] == "ing":
    list3.append(word.partition("ing"))
  elif word[-7:] == "isation":
    list3.append(word.partition("isation"))
  else:
    list3.append(word)

print(list3)
```

```python
list4 = []
for element in list3:
  if type(element) is tuple:
    list4.append(element[0])
    list4.append(element[1])
  else:
    list4.append(element)
print(list4)
```
